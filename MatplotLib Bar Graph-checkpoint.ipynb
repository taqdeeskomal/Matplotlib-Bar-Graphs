```python
t=np.arange(0,10)
amp1=t
amp2=t**2
plt.plot(t,amp1,color='k')
plt.plot(t,amp2,color='r')
plt.xlabel("time")
plt.ylabel("amplitude")
plt.show()
```

```python
t=np.arange(0,10)
amp1=t
amp2=t**2
plt.plot(t,amp1,color='k')
plt.plot(t,amp2,color='r')
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("graph")
plt.show()
```

```python
t=np.arange(0,10)
amp1=t
amp2=t**2
plt.plot(t,amp1,color='k')
plt.plot(t,amp2,color='r')
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("graph")
plt.show()
```

```python
t=np.arange(0,10)
amp1=t
amp2=t**2
plt.plot(t,amp1,color='k')
plt.plot(t,amp2,color='r')
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("graph",loc="right")
plt.legend(["y=t","y=t^2"])
plt.grid()
plt.show()
```

```python
t=np.arange(0,10)
amp1=t
amp2=t**2
plt.plot(t,amp1,color='k')
plt.plot(t,amp2,color='r')
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("graph")
plt.grid(axis='x')
plt.show()
```

```python
t=np.arange(0,10)
amp1=t
amp2=t**2
plt.plot(t,amp1,color='k')
plt.plot(t,amp2,color='r')
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("graph")
plt.grid(color="r",linestyle="dotted",linewidth=5)
plt.show()
```
